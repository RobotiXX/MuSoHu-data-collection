```python
import os
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support import expected_conditions as EC

def activate_page(driver, page_number):
    # Find the pagination element
    paginator = driver.find_element(By.CLASS_NAME, "ui-paginator-pages")

    # # Find the link/button for the desired page number
    # page_link =  paginator.find_element(By.LINK_TEXT, str(page_number)) # Assuming page numbers are text links
    print(page_number+1)

    # # Click the link/button to activate the page
    nextpage = '//*[@id="datasetForm:tabView:filesTable_paginator_bottom"]/a[3]'
    # page3 = '//*[@id="datasetForm:tabView:filesTable_paginator_bottom"]/span/a[3]'
    button = driver.find_element(By.XPATH, nextpage)
    # //*[@id="datasetForm:tabView:filesTable_paginator_bottom"]/span/a[5]
    #//*[@id="ToolList"]/ul/li[1]/div/a/p/span[1]/span
    # Click the button
    button.click()
    time.sleep(10)
    # Find all <a> elements on the page
    all_links = driver.find_elements(By.TAG_NAME,"a")
    # print(all_links)
    # Extract href attributes from each <a> element
    for link in all_links:

        href = link.get_attribute("href")
        if href is None:
            continue
        if "https://dataverse.orc.gmu.edu/file.xhtml?" in href:
            if ".bag" not in link.text.strip():
                continue
            print(link.text.strip()," ; ",href[40:-12])
            os.system(
                    "curl -L -O -J https://DATAVERSE.orc.gmu.edu/api/access/datafile/:persistentId"+href[40:-12])
```

```python
from bs4 import BeautifulSoup

def find_bag_file_names(html_file_path, target_href_pattern):
    with open(html_file_path, "r", encoding="utf-8") as file:
        html_content = file.read()

    soup = BeautifulSoup(html_content, "html.parser")

    # Find <a> tags with href attribute containing the specified pattern
    a_tags = soup.select("a[href*='" + target_href_pattern + "']")

    # Extract text content (file names) from the <a> tags
    href_and_file_names = [(a_tag["href"], a_tag.text.strip()) for a_tag in a_tags]

    return href_and_file_names
```

```python
import os
# Example usage
html_file_path = "../Socially Compliant Navigation Dataset (SCAND) - Texas Robotics.html"  # Replace with the path to your HTML file
target_href_pattern = "https://dataverse.tdl.org/file.xhtml?"

# Find names of .bag files within <a> tags with href attribute containing the specified pattern
href_and_file_names = find_bag_file_names(html_file_path, target_href_pattern)

# Print the found file names
for href_name, file_name in href_and_file_names:
    print(href_name.split("&")[0].split("=")[1], file_name)
    if "Spot" in file_name:
      if ".avi" in file_name or ".mp4" in file_name or ".bag" in file_name:
        os.system(
                  "curl -L -O -J https://dataverse.tdl.org/api/access/datafile/"+href_name.split("&")[0].split("=")[1])
```
